# Lab 04: Statistical Analysis and Data Visualization

### Objective
Perform various statistical calculations and visualizations on randomly generated data.

## Task 1: Generate Random Data and Calculate the Mean
Generate random data with a sample size of 100 from a normal distribution with mean=10 and std=1.

**Formula for Mean:**
$$
 \text{Mean} = \frac{\sum_{i=1}^N x_i}{N}
$$

The mean (average) of a dataset is calculated by summing all values and dividing by the total count.

**Function Syntax:** `np.mean(data)`

We'll visualize this using a histogram and mark the mean on the plot.

````python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data
data = np.random.normal(loc=10, scale=1, size=100)

# Calculate mean
mean = np.mean(data)
print("Mean:", mean)

# Plot histogram
plt.hist(data, bins=10, color='skyblue', edgecolor='black')
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label=f"Mean: {mean:.2f}")
plt.legend()
plt.title('Data Distribution with Mean')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
````

## Task 2: Calculate the Median
Using the same dataset, calculate the median and visualize it on the histogram.

**Formula for Median:**
- If the data is sorted and has an odd number of values, the median is the middle value.
- If the data is sorted and has an even number of values, the median is the average of the two middle values.

**Function Syntax:** `np.median(data)`

````python
# Calculate median
median = np.median(data)
print("Median:", median)

# Plot histogram
plt.hist(data, bins=10, color='lightcoral', edgecolor='black')
plt.axvline(median, color='blue', linestyle='dashed', linewidth=2, label=f"Median: {median:.2f}")
plt.legend()
plt.title('Data Distribution with Median')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
````

## Task 3: Calculate the Mode
Calculate the mode(s) of the dataset and mark it on the histogram.

**Formula for Mode:**
The mode is the value(s) that appears most frequently in the dataset.

**Function Syntax:** `scipy.stats.mode(data)`

````python
from scipy.stats import mode

from scipy.stats import mode

# Calculate mode
mode_result = mode(data)
mode_value = mode_result.mode
print("Mode:", mode_value)

# Plot histogram
plt.hist(data, bins=10, color='lightgreen', edgecolor='black')
plt.axvline(mode_value, color='purple', linestyle='dashed', linewidth=2, label=f"Mode: {mode_value}")
plt.legend()
plt.title('Data Distribution with Mode')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
````

## Task 4: Range (Minimum and Maximum)
Calculate the minimum and maximum values to find the range and visualize using a boxplot.

**Formula for Range:**
$$
 \text{Range} = \text{Maximum} - \text{Minimum}
$$

**Function Syntax:** `np.min(data)` and `np.max(data)`

````python
# Calculate min and max
min_value = np.min(data)
max_value = np.max(data)
print("Minimum:", min_value)
print("Maximum:", max_value)

# Plot boxplot
plt.boxplot(data)
plt.scatter([1, 1], [min_value, max_value], color='red', label='Min/Max')
plt.legend()
plt.title('Data Distribution with Min and Max')
plt.ylabel('Value')
plt.show()
````

## Task 5: Variance
Calculate the variance and plot the data distribution, displaying the variance as text.

**Formula for Variance:**
$$
 \text{Variance} = \frac{\sum_{i=1}^N (x_i - \text{Mean})^2}{N}
$$

**Function Syntax:** `np.var(data)`

````python
# Calculate variance
variance = np.var(data)
print("Variance:", variance)

# Plot histogram with variance
plt.hist(data, bins=10, color='lightblue', edgecolor='black')
plt.text(0.5, 0.5, f"Variance: {variance:.2f}", transform=plt.gca().transAxes)
plt.title('Data Distribution with Variance')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
````

## Task 6: Standard Deviation
Calculate the standard deviation and annotate 1, 2 standard deviations on the histogram.

**Formula for Standard Deviation:**
$$
 \text{Standard Deviation} = \sqrt{\text{Variance}}
$$

**Function Syntax:** `np.std(data)`

````python
# Calculate standard deviation
std_dev = np.std(data)
print("Standard Deviation:", std_dev)

# Plot histogram with standard deviations
plt.hist(data, bins=10, color='lightgray', edgecolor='black')
for i in range(1, 3):
    plt.axvline(mean + i * std_dev, color='orange', linestyle='dotted', linewidth=1, label=f"+{i} Std Dev")
    plt.axvline(mean - i * std_dev, color='orange', linestyle='dotted', linewidth=1, label=f"-{i} Std Dev")
plt.legend()
plt.title('Data Distribution with Standard Deviations')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
````

## Task 7: Calculate Q1, Q3, and IQR
Compute the first quartile (Q1), third quartile (Q3), and the interquartile range (IQR). Visualize the data using a boxplot showing quartiles and IQR.

**Formulas:**
- **Q1**: The 25th percentile of the data.
- **Q3**: The 75th percentile of the data.
- **IQR**: $$ \text{IQR} = Q3 - Q1 $$

**Function Syntax:**
```python
np.percentile(data, 25)  # Q1
np.percentile(data, 75)  # Q3
```


````python
# Calculate Q1, Q3, and IQR
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1
print("Q1:", Q1)
print("Q3:", Q3)
print("IQR:", IQR)

# Plot boxplot with quartiles
plt.boxplot(data)
plt.axhline(y=Q1, color='green', linestyle='dashed', label=f"Q1: {Q1}")
plt.axhline(y=Q3, color='purple', linestyle='dashed', label=f"Q3: {Q3}")
plt.legend()
plt.title('Boxplot with Quartiles and IQR')
plt.ylabel('Value')
plt.show()
````

## Task 8: Generate Skewed Data, Calculate Mean and Median
Generate two skewed datasets (left-skewed and right-skewed), calculate the mean and median, and plot histograms marking mean, median, and mode.

**Approach**:
1. Generate skewed data.
2. Calculate mean and median.
3. Use the same histogram plotting approach from earlier tasks, marking each measure.

**Function Syntax for Generating Skewed Data:**
```python
np.random.exponential(scale, size)  # Right skew
-np.random.exponential(scale, size)  # Left skew
```


````python
# Generate right-skewed data
right_skewed_data = np.random.exponential(scale=2, size=100)
# Generate left-skewed data
left_skewed_data = -np.random.exponential(scale=2, size=100)

# Calculate mean and median for right-skewed data
right_mean = np.mean(right_skewed_data)
right_median = np.median(right_skewed_data)
print("Right Skewed Data - Mean:", right_mean)
print("Right Skewed Data - Median:", right_median)

# Plot histogram for right-skewed data
plt.hist(right_skewed_data, bins=10, color='lightblue', edgecolor='black')
plt.axvline(right_mean, color='red', linestyle='dashed', label=f"Mean: {right_mean:.2f}")
plt.axvline(right_median, color='blue', linestyle='dashed', label=f"Median: {right_median:.2f}")
plt.legend()
plt.title('Right-Skewed Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

# Calculate mean and median for left-skewed data
left_mean = np.mean(left_skewed_data)
left_median = np.median(left_skewed_data)
print("Left Skewed Data - Mean:", left_mean)
print("Left Skewed Data - Median:", left_median)

# Plot histogram for left-skewed data
plt.hist(left_skewed_data, bins=10, color='lightcoral', edgecolor='black')
plt.axvline(left_mean, color='red', linestyle='dashed', label=f"Mean: {left_mean:.2f}")
plt.axvline(left_median, color='blue', linestyle='dashed', label=f"Median: {left_median:.2f}")
plt.legend()
plt.title('Left-Skewed Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
````

## Task 9: Calculate Z-Scores
Calculate the Z-scores of the data and plot a histogram of the Z-scores, comparing it to a standard normal distribution.

**Formula for Z-score:**
$$
 Z = \frac{X - \text{Mean}}{\text{Standard Deviation}}
$$

**Function Syntax for Z-scores:**
```python
(data - np.mean(data)) / np.std(data)
```


````python
# Calculate Z-scores
z_scores = (data - np.mean(data)) / np.std(data)
print("First 10 Z-scores:", z_scores[:10])

# Plot histogram of Z-scores
plt.hist(z_scores, bins=10, color='skyblue', edgecolor='black')
plt.title('Histogram of Z-scores')
plt.xlabel('Z-score')
plt.ylabel('Frequency')
plt.axvline(0, color='red', linestyle='dashed', label='Mean (0)')
plt.axvline(1, color='orange', linestyle='dashed', label='+1 Std Dev')
plt.axvline(-1, color='orange', linestyle='dashed', label='-1 Std Dev')
plt.legend()
plt.show()
````

## Task 10: Calculate Q1, Q3, IQR, and Identify Outliers
Calculate Q1, Q3, and IQR, then identify outliers (values outside 1.5*IQR from Q1 or Q3) and visualize using a scatter plot or boxplot.

**Formula for Outliers:**
Any value outside the range:
$$
 [Q1 - 1.5 \times \text{IQR}, Q3 + 1.5 \times \text{IQR}]
$$
is considered an outlier.

**Function Syntax:**
```python
outliers = data[(data < Q1 - 1.5 * IQR) | (data > Q3 + 1.5 * IQR)]
```


````python
# Identify outliers
outliers = data[(data < Q1 - 1.5 * IQR) | (data > Q3 + 1.5 * IQR)]
print("Outliers:", outliers)

# Plot boxplot with outliers
plt.boxplot(data, flierprops=dict(marker='o', color='red', markersize=8))
plt.title('Boxplot with Outliers')
plt.ylabel('Value')
plt.show()
````
