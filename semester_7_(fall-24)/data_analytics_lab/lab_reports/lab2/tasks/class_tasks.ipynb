```python
import matplotlib.pyplot as plt
import numpy as np
```

Task 2: Create a Pie chart for Gender and interpret it.

```python
# Data for gender distribution, assuming realistic values
gender_labels = ['Male', 'Female']
gender_counts = [4, 4]  # Equal distribution of 4 males and 4 females, from the document

# Create the pie chart
plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=gender_labels, autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightcoral'])
plt.title('Gender Distribution')

# Display the chart
plt.show()
```

Task 3: Create a Bar chart for Age and interpret it. 

```python
# Data for age distribution, assuming realistic counts for ages between 5 and 60 years
ages = np.arange(5, 61, 5)  # Age categories (5, 10, 15, ..., 60)
age_counts = np.random.randint(1, 10, size=len(ages))  # Random counts for each age category

# Create the bar chart
plt.figure(figsize=(10, 6))
plt.bar(ages, age_counts, color='skyblue')
plt.xlabel('Age')
plt.ylabel('Number of People')
plt.title('Age Distribution')
plt.xticks(ages)

# Display the chart
plt.show()
```

Task 4: Create a Bar chart for Domicile and interpret it.

```python
# Data for domicile, assuming realistic counts for the cities
cities = [
    "Bannu", "Charsadda", "Chitral", "DI Khan", "Dir", "DIR", "FR Kohat", 
    "Hangu", "Karak", "Kashmir", "Khyber", "Kohat", "Kurran", "Mardan", 
    "Nangarhar", "Orakzai Agency", "Peshawar", "Swabi", "Waziristan"
]
city_counts = np.random.randint(1, 10, size=len(cities))  # Random counts for each city

# Create the bar chart
plt.figure(figsize=(10, 8))
plt.barh(cities, city_counts, color='lightgreen')
plt.xlabel('Number of People')
plt.ylabel('Cities')
plt.title('Domicile Distribution by Cities')

# Display the chart
plt.show()
```
