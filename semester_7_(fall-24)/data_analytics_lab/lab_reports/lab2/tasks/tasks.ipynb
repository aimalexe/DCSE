In this lab, we will create frequency tables and visualizations for Gender, Age, Domicile, and CGPA using Python's NumPy and Matplotlib libraries.

## Task 1: Frequency Tables
Create Frequency Tables for Gender, Age, Domicile, and CGPA.

We will generate random data for each variable and compute the frequency tables.

```python
import numpy as np
import pandas as pd

# Sample data for Gender, Age, Domicile, and CGPA
gender = np.random.choice(['Male', 'Female'], size=100)
age = np.random.choice([20, 21, 22, 23, 24], size=100)
domicile = np.random.choice([
    'Bannu', 'Charsadda', 'Chitral', 'DI Khan', 'Dir', 'Kohat',
    'Mardan', 'Orakzai Agency', 'Peshawar', 'Swabi', 'Waziristan'
], size=100)
cgpa = np.round(np.random.uniform(2.5, 3.8, size=100), 2)

# Create frequency tables
gender_freq = pd.value_counts(gender).to_frame(name='Frequency')
age_freq = pd.value_counts(age).to_frame(name='Frequency')
domicile_freq = pd.value_counts(domicile).to_frame(name='Frequency')
cgpa_freq = pd.value_counts(cgpa).sort_index().to_frame(name='Frequency')

print("Frequency Table for Gender")
print(gender_freq)

print("\nFrequency Table for Age")
print(age_freq)

print("\nFrequency Table for Domicile")
print(domicile_freq)

print("\nFrequency Table for CGPA")
print(cgpa_freq)
```

## Task 2: Pie Chart for Gender
Create a Pie chart for Gender and interpret it.

We will use Matplotlib's `pie` function to create a pie chart of the Gender distribution.

```python
import matplotlib.pyplot as plt

# Pie chart for Gender
gender_counts = gender_freq['Frequency']
plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Gender Distribution')
plt.show()
```

## Task 3: Bar Chart for Age
Create a Bar chart for Age and interpret it.

We will use Matplotlib's `bar` function to create a bar chart showing the distribution of ages.

```python
# Bar chart for Age
plt.figure(figsize=(8, 6))
plt.bar(age_freq.index, age_freq['Frequency'], color='skyblue')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')
plt.show()
```

## Task 4: Bar Chart for Domicile
Create a Bar chart for Domicile and interpret it.

Using `bar` function, we will plot the distribution of students' domiciles.

```python
# Bar chart for Domicile
plt.figure(figsize=(10, 6))
plt.bar(domicile_freq.index, domicile_freq['Frequency'], color='lightgreen')
plt.xlabel('Domicile')
plt.ylabel('Frequency')
plt.title('Domicile Distribution')
plt.xticks(rotation=45)
plt.show()
```

## Task 5: Histogram for CGPA
Create a Histogram for CGPA and interpret it.

We will use Matplotlib's `hist` function to plot a histogram showing the distribution of CGPA scores.

```python
# Histogram for CGPA
plt.figure(figsize=(8, 6))
plt.hist(cgpa, bins=10, color='salmon', edgecolor='black')
plt.xlabel('CGPA')
plt.ylabel('Frequency')
plt.title('CGPA Distribution')
plt.show()
```
